```python
import pandas as pd
import shutil
import os.path
```

```python
# Read in the original database
df = pd.read_csv('maestro-v3.0.0/maestro-v3.0.0.csv')
```

```python
print(df.columns)
```

```python
# Create a new column with the number of instances that a song with the same name appears
df['song_count'] = df['canonical_title'].map(df['canonical_title'].value_counts())
#print(df)
```

```python
# Filters through the original dataframe to only include songs that appear more than 10 times
# so that we have suffi
filtered = df[df.song_count > 10]
filtered.reset_index(inplace=True)
#print(filtered)
```

```python
for index, row in filtered.iterrows():
    if not os.path.exists(f'references/{row['canonical_title']}'):
        os.mkdir(f'references/{row['canonical_title']}')

    midi_path = os.path.join('maestro-v3.0.0', row['midi_filename'])
    shutil.copy(midi_path, f'references/{row['canonical_title']}')
    audio_path = os.path.join('maestro-v3.0.0', row['audio_filename'])
    shutil.copy(audio_path, f'references/{row['canonical_title']}')
```
